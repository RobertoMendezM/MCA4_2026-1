```python
import numpy as np
import matplotlib.pyplot as plt
```

# EDO de 1° orden $\frac{dy}{dx}=f(x)$

```python
"""
Primero crearemos una función que nos grafique de manera automatica el campo direccional y las curvas de solución. Utilizaremos los siguientes parámetros:
fx: Función f(x)
yx: Solución a la ec. diferencial y(x,c), la construimos con dos variables donde x es la variable independiente y c es los distintos valores de la constante
x_val: Lista de dos valores que representan el intervalo de solución en el eje x
y_val: Lista de dos valores que representan el intervalo de solución en el eje y
step: Cantidad de pedazos en los que dividiremos ambos intervalos
c_val: Lista que contiene los distintos valores de constantes para las curvas de solución
"""

def edo_basica(fx,yx,x_val,y_val,step,c_val):
  x=np.linspace(x_val[0],x_val[1],step)
  y=np.linspace(y_val[0],y_val[1],step)
  X,Y=np.meshgrid(x,y)

  #Create normal unit vectors
  dx=f(X)
  dy=np.ones_like(X)

  length=np.sqrt(dx**2+dy**2)
  dx,dy=dx/length,dy/length

  #Generar puntos para las curvas de solucion
  if yx=='None':
    plt.quiver(X, Y, dy, dx, color='black', headlength=3)
    plt.title("Direction Field and Solution Curves")
    plt.xlim(x_val[0],x_val[1])
    plt.ylim(y_val[0],y_val[1])
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend(loc='best')
    plt.grid(True)
    plt.show()
  else:
    x_curves=np.linspace(x_val[0],x_val[1],step)
    for c in c_val:
      y_curves=yx(x_curves,c)
      plt.plot(x_curves,y_curves)
    plt.quiver(X, Y, dy, dx, color='black', headlength=3)
    plt.title("Direction Field and Solution Curves")
    plt.xlim(x_val[0],x_val[1])
    plt.ylim(y_val[0],y_val[1])
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend(loc='best')
    plt.grid(True)
    plt.show()
```

## ¿Cómo dibujar campos de dirección?

## Ejemplo 1. $y'=x^2$

```python
def f(x):
  return x**2
def y(x,c):
  return (x**3/3)+c
x_val=[-5,5]
y_val=[-5,5]
step=30
c_val=[-2,-1,0.5,1,2]
edo_basica(f,y,x_val,y_val,step,c_val)
```

## Ejemplo 2. Caída libre: $y''=-g$

```python
grav=9.8
def f(x):
    return -grav
def y(x,c):
  return -grav*x+c
x_val=[-5,5]
y_val=[-5,5]
step=30
c_val=[-2,-1,0.5,1,2]
edo_basica(f,y,x_val,y_val,step,c_val)
```

# EDO 1° Orden $\frac{dy}{dx}=f(y)$ (autónomas)

```python
"""
Primero crearemos una función que nos grafique de manera automatica el campo direccional y las curvas de solución. Utilizaremos los siguientes parámetros:
fy: Función f(y)
yx: Solución a la ec. diferencial y(x,c), la construimos con dos variables donde x es la variable independiente y c es los distintos valores de la constante
x_val: Lista de dos valores que representan el intervalo de solución en el eje x
y_val: Lista de dos valores que representan el intervalo de solución en el eje y
step: Cantidad de pedazos en los que dividiremos ambos intervalos
c_val: Lista que contiene los distintos valores de constantes para las curvas de solución
sol_eq: Lista que contiene las soluciones de equilibrio del sistema, es decir y(x)=c donde c es un punto crítico de la ec. diferencial.
"""

def edo_autonoma(fy,yx,x_val,y_val,step,c_val,sol_eq):
  x=np.linspace(x_val[0],x_val[1],step)
  y=np.linspace(y_val[0],y_val[1],step)
  X,Y=np.meshgrid(x,y)

  dy=f(Y)
  dx=np.ones_like(X)

  length=np.sqrt(dx**2+dy**2)
  dx,dy=dx/length,dy/length

  #Generar puntos para las curvas de solucion
  if yx=='None':
    for sol in sol_eq:
      plt.axhline(y=sol, linestyle='--')
    plt.quiver(X, Y, dx, dy, color='black', headlength=3)
    plt.title("Direction Field and Solution Curves")
    plt.xlim(x_val[0],x_val[1])
    plt.ylim(y_val[0],y_val[1])
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend(loc='best')
    plt.grid(True)
    plt.show()
  else:
    x_curves=np.linspace(x_val[0],x_val[1],step)
    for c in c_val:
      y_curves=yx(x_curves,c)
      plt.plot(x_curves,y_curves)
    # Graficar las soluciones de equilibrio
    for sol in sol_eq:
      plt.axhline(y=sol, linestyle='--')
    plt.quiver(X, Y, dx, dy, color='black', headlength=3)
    plt.title("Direction Field and Solution Curves")
    plt.xlim(x_val[0],x_val[1])
    plt.ylim(y_val[0],y_val[1])
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend(loc='best')
    plt.grid(True)
    plt.show()
```

## Ejemplo 1: $\frac{dy}{dx}=y(1-y)$ (ecuación logística)

```python
def f(y):
  return y*(1-y)
def y(x,c):
  return (c * np.exp(x)) / (1 + c * np.exp(x))
x_val=[-5,5]
y_val=[-5,5]
step=30
c_val=[-2,-1,0.5,1,2]
sol_eq=[0,1]
edo_autonoma(f,y,x_val,y_val,step,c_val,sol_eq)
```

## $\frac{dy}{dx}=y^{2}-3y-10$

```python
def f(y):
  return y**2-3*y-10
def y(x,c):
  return (5+2*c * np.exp(7*x)) / (1 - c * np.exp(7*x))
x_val=[-2,2]
y_val=[-5,10]
step=30
c_val=[-2,-1,0.5,1,2]
sol_eq=[-2,5]
edo_autonoma(f,y,x_val,y_val,step,c_val,sol_eq)
```
